## 2. Quality Assurance Checks
This section executes a high-level statistical overview (`describe`), checks for missing data points (`nulls`), and performs basic assertion checks on the logical structure of market prices (e.g., ensuring a trading day's High price is mathematically sound relative to its Low and Close prices).

```python
# Quality Check: Understand the data distribution and identify any missing values

# Generate descriptive statistics summarizing central tendency, dispersion, and shape of the dataset
display(df.describe())

# Check for missing values in each column and calculate the total count of nulls per column
print("\n--- Null Value Check ---")
print(df.isnull().sum())

# Verify price logic: The 'High' price must always be greater than or equal to both 'Low' and 'Close' prices
# Create a new DataFrame containing any rows where this logic is violated
logic_errors = df[(df['High_Price'] < df['Low_Price']) | (df['High_Price'] < df['Close_Price'])]
if not logic_errors.empty:
    print("\n--- Logic Errors Found ---")
    display(logic_errors)
else:
    print("\nNo price logic errors found.")
```
